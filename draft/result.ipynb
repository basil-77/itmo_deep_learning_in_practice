```python
import matplotlib.pyplot as plt

# Данные для первого источника
x_values1 = v5['                  epoch']
y_values1 = v5['   metrics/precision(B)']

# Данные для второго источника
x_values2 = v8['                  epoch']
y_values2 = v8['   metrics/precision(B)']

# Построение линейного графика для первого источника
plt.plot(x_values1, y_values1, label='YOLOv5', marker='o')

# Построение линейного графика для второго источника
plt.plot(x_values2, y_values2, label='YOLOv8', marker='o')

# Добавление заголовка и меток осей
plt.title('Сравнение двух моделей')
plt.xlabel('Количество эпох')
plt.ylabel('precision')

# Добавление легенды
plt.legend()

# Отображение графика
plt.show()
```
